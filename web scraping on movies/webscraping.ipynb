```python
url="https://www.flipkart.com/search?q=AC&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text)
soup
```

```python
b=[]
for i in soup.find_all("div",class_="_4rR01T"):
        regex=re.findall("^[A-Za-z]+",i.text)
        if regex:
            b.append(regex[0])
        else:
            b.append(np.nan)
```
